# Split texts to list by "|"

```python
df['categories_list'] = (
    df['categories']
    .fillna('')
    .astype(str)
    .apply(lambda s: [x.strip) for x in s.split('|') if x.strip()])
)


df['keyword_list'] = (
    df['matched_keywords']
    .fillna('')
    .astype(str)
    .apply(lambda s: [x.strip() for x in s.split('|') if x.strip()])
)

df = df[['date','description','categories_list', 'keyword_list']]
df.columns = ['date','text', 'category' , 'keyword']
```
